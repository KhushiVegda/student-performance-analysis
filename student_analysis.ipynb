# Student Performance Data Analysis

This project analyzes the **Student Performance Dataset** using Python.

The goal of this project is to understand how different factors such as
study time, gender, and other attributes affect the **final grade (G3)** of students.

The dataset contains information about students including:
- Gender
- Study time
- Family background
- School activities
- Grades (G1, G2, G3)

We perform **data cleaning, analysis, and visualization** to discover patterns in student performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Dataset Loading**
### Step 1: Load Dataset

In this step, we load the dataset using the **pandas library**.

The dataset file **student-mat.csv** contains student information and academic performance data.

We use `pd.read_csv()` to load the dataset into a DataFrame for further analysis.

```python
# Load dataset (important: dataset uses semicolon separator)
df = pd.read_csv("student-mat.csv", sep=';')

# Display first 5 rows
df.head()
```

**Data Exploration & Cleaning**
## Step 2: Explore and Clean Data

Before performing analysis, it is important to understand and clean the dataset.

The following steps were performed:

• Checked the dataset shape to know the number of rows and columns  
• Inspected column data types using `.dtypes`  
• Checked for missing values using `.isnull()`  
• Checked for duplicate records and removed them if present  

Cleaning the data ensures accurate analysis results.

```python
# Number of rows and columns
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

**Data Analysis**
## Step 3: Data Analysis

In this step, we answer important questions about student performance.

The following analyses were performed:

• Calculated the **average final grade (G3)**  
• Counted how many students scored **above 15**  
• Checked if **study time is correlated with performance**  
• Compared **average grades between male and female students**

This helps us understand which factors influence academic performance.

```python
# Calculate average final grade
average_grade = df['G3'].mean()

print("Average Final Grade:", average_grade)
```

```python
high_score_students = df[df['G3'] > 15]

print("Number of students scoring above 15:", len(high_score_students))
```

```python
correlation = df['studytime'].corr(df['G3'])

print("Correlation between study time and final grade:", correlation)
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

print(gender_avg)
```

**Data Visulization**
## Step 4: Data Visualization

Visualization helps us better understand the data patterns.

The following graphs were created:

• **Histogram of grades** to show grade distribution  
• **Scatter plot** to analyze relationship between study time and grades  
• **Bar chart** to compare male and female average scores  

These visualizations make the analysis easier to interpret.

```python
plt.figure(figsize=(8,5))

plt.hist(df['G3'], bins=10)

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['studytime'], df['G3'])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")

plt.show()
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

plt.figure(figsize=(6,5))

gender_avg.plot(kind='bar')

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")

plt.show()
```

## Conclusion

From this analysis we observed the following:

• The dataset provides insights into student academic performance.  
• The average final grade helps understand overall student performance.  
• Some students scored above 15, indicating strong academic achievement.  
• Study time has some relationship with student performance.  
• Gender comparison shows differences in average scores.

Data analysis and visualization help identify patterns that can improve understanding of student learning behavior.
